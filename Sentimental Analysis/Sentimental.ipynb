```python
from transformers import pipeline
classifier = pipeline("text-classification",model='bhadresh-savani/distilbert-base-uncased-emotion', return_all_scores=True)

```

```python
prediction = classifier("In GTA V, if you force Trevor to listen to a radio station he doesn’t like too many times, he will get increasingly aggravated and will override you and switch the stations himself", )
print(prediction)
```

```python
import pandas as pd
from pywikihow import RandomHowTo, HowTo
from transformers import pipeline

#Resume
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
#WikiHow
how_to = HowTo("https://www.wikihow.com/Study")
```

```python
#Read data
#post = pd.read_excel("WikiHow.xlsx", index_col=0)
```

```python
#Create Dataset
data = {"Title" : ["Titulo prueba"], "Description": ["Description prueba"], "Summary": ["Summary prueba"], "SummaryShorter": ["SummaryShorter prueba"]}
post = pd.DataFrame(data)
post.to_excel("SummaryOfTheSummary.xlsx")
```

```python
for step in how_to.steps:
  dat = step.as_dict()
  resume = summarizer(dat["description"], max_length=150, min_length=50, do_sample=False)
  resume2 = summarizer(dat["description"], max_length=80, min_length=25, do_sample=False)
  prueba = {"Title":dat["summary"], "Description": dat["description"], "Summary": resume[0]["summary_text"], "SummaryShorter": resume2[0]["summary_text"]}
  post = post.append(prueba, ignore_index=True)
  # print(dat)
```

```python
 ##Summary of the Summary
 
 prep = ""
 sett = ""
 using = ""
 using2 = ""
 stud = ""
 stud2 = ""

 #Dividimos por secciones del wikihow, algunos en dos porque son muy largos y no caben en el summarizer
 for step in how_to.steps:
   dat = step.as_dict()
   if dat["number"] < 6:
     prep = prep + dat["description"]
   elif dat["number"] < 11:
      sett = sett + dat["description"]
   elif dat["number"] < 15:
      using = using + dat["description"]
   elif dat["number"] < 19:
      using2 = using2 + dat["description"]
   elif dat["number"] < 24:
      stud = stud + dat["description"]
   elif dat["number"] < 27:
      stud2 = stud2 + dat["description"]

#Resumimos todos
prepSum = summarizer(prep, max_length=600, min_length=50, do_sample=False)
prepSum2 = summarizer(prep, max_length=300, min_length=50, do_sample=False)
settSum = summarizer(sett, max_length=600, min_length=50, do_sample=False)
settSum2 = summarizer(sett, max_length=300, min_length=50, do_sample=False)
usingSum = summarizer(using, max_length=1000, min_length=50, do_sample=False)
using2Sum = summarizer(using2, max_length=1000, min_length=50, do_sample=False)
studSum = summarizer(stud, max_length=1000, min_length=50, do_sample=False)
stud2Sum = summarizer(stud2, max_length=1000, min_length=50, do_sample=False)

#Juntamos los partidos y los volvemos a resumir para que se queden como los otros
using = usingSum[0]["summary_text"] + using2Sum[0]["summary_text"]
stud = studSum[0]["summary_text"] + stud2Sum[0]["summary_text"]
usingSum = summarizer(using, max_length=600, min_length=50, do_sample=False)
usingSum2 = summarizer(using, max_length=300, min_length=50, do_sample=False)
studSum = summarizer(stud, max_length=600, min_length=50, do_sample=False)
studSum2 = summarizer(stud, max_length=300, min_length=50, do_sample=False)

#Juntamos todos y hacemos el resumen final
tot = prepSum[0]["summary_text"] + settSum[0]["summary_text"] + usingSum[0]["summary_text"] + studSum[0]["summary_text"]
totSum = summarizer(tot, max_length=600, min_length=50, do_sample=False)
totSum2 = summarizer(tot, max_length=300, min_length=50, do_sample=False)
```

```python
#Añadimos los resumenes grandes al excel
post = post.append({"Title":"Preparing to Study", "Description": "", "Summary": prepSum[0]["summary_text"], "SummaryShorter": prepSum2[0]["summary_text"]}, ignore_index=True)
post = post.append({"Title":"Setting up your Study Space", "Description": "", "Summary": settSum[0]["summary_text"], "SummaryShorter": settSum2[0]["summary_text"]}, ignore_index=True)
post = post.append({"Title":"Using Effective Study Techniques", "Description": "", "Summary": usingSum[0]["summary_text"], "SummaryShorter": usingSum2[0]["summary_text"]}, ignore_index=True)
post = post.append({"Title":"Studying More Efficiently", "Description": "", "Summary": studSum[0]["summary_text"], "SummaryShorter": studSum2[0]["summary_text"]}, ignore_index=True)
post = post.append({"Title":"Study", "Description": "", "Summary": totSum[0]["summary_text"], "SummaryShorter": totSum2[0]["summary_text"]}, ignore_index=True)
```

```python
#SAVE
post.to_excel("SummaryOfTheSummary.xlsx")
```

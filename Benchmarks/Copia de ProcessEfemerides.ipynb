```python
from torch.utils.data.dataloader import T
import pandas as pd
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
from urllib.request import Request


for i in range(1,32):
  #specify URL
  url = "https://thisdayintechhistory.com/1/{}/".format(i)
  print(url)

  #get the HTML from the page and assign it to the html variable
  html = urlopen(url)
  #req = Request(url, headers={'User-Agent': 'Mozilla/6.0'})
  #html = urlopen(req)

  #parse the HTML and assign it to the variable 'soup'
  soup = bs(html.read(), 'html.parser')

  #get the text of the joke and assign to the variable "chiste"
  joke = soup.find("div", class_="fl-post-content clearfix").text
  print(joke)
  #Clean the data
  date = ""
  year = ""
  joke = joke[1:]
  flag = False
  for n in joke:
    if n == ",":
      flag = True
      continue
    if n == "\n":
      break
    if flag == False:
      date += n
    if flag == True:
      year += n
    joke = joke[1:]
  print(date)
  print(year)
  print(joke)
  efemeride = ""
  for n in joke[2:]:
    if n == "\n":
      break
    efemeride += n
  print(efemeride)

  summary = summarizer(efemeride, max_length=50, min_length=30, do_sample=False)
  trans = translated.translate(summary[0]["summary_text"])
  dat = {"Efeméride": efemeride, "Resumen":summary, "Relación Semántica": similar(efemeride, summary)}
  post = post.append(dat, ignore_index=True)
```

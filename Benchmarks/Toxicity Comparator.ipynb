```python
from detoxify import Detoxify

for x in range(10000):
  max = 0
  text = data["comment_text"][x]
  toxi = Detoxify('original').predict(str(text))
  #If it contains any type of toxicity
  for v in iter(toxi.values()):
    if v > max:
      max = v
  if (data["toxic"][x] == 1) or (data["severe_toxic"][x] == 1) or (data["obscene"][x] == 1) or (data["threat"][x] == 1) or (data["insult"][x] == 1) or (data["identity_hate"][x] == 1):
    toxic = 1
  else:
    toxic = 0
  if (max > 0.7) and (toxic == 1):
    resultado = 1
  elif (max < 0.7) and (toxic == 0):
    resultado = 1
  else:
    resultado = 0

  if (max < 0.7) and (toxic == 1):
    xx = 1
  else:
    xx = 0

  
  prueba = {"Fact": text, "Toxicidad Prueba":max, "Toxicidad Real": toxic, "Fibiabilidad":resultado, "Falsos Negativos": xx}
  post = post.append(prueba, ignore_index=True)
```

```python
#SAVE
post.to_excel("/content/drive/MyDrive/Trabajo Fin de Grado/Final-Degree-Project/Benchmarks/ToxicityBenchmark.xlsx")
```

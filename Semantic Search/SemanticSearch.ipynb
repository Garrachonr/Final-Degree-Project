```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-mpnet-base-v2')
```

```python
first = model.encode("Un dia vi una vaca vestida de uniforme")
second = model.encode("Las vacas comen mucho pasto al lado del uniforme")
```

```python
print("Similarity:", util.dot_score(first, second))
```

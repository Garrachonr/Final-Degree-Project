```python
#Importar librerias y tal

from parrot import Parrot
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch
import warnings
warnings.filterwarnings("ignore")
```

```python
#Frase a parafrasear
phrase = "Some concepts or classes will come to you more naturally, so you can study those more quickly"
```

```python
#Generar una semilla para la recreación de las frases
def random_state(seed):
  torch.manual_seed(seed)
  if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)

random_state(1234)
```

```python
#PARROT
parrot = Parrot(model_tag="prithivida/parrot_paraphraser_on_T5", use_gpu=False)
```

```python
#Parafreamos y sacamos 4 ejemplos con parrot
#https://github.com/PrithivirajDamodaran/Parrot_Paraphraser
print("-"*100)
print("Input_phrase: ", phrase)
print("-"*100)
para_phrases = parrot.augment(input_phrase=phrase, max_return_phrases = 5, do_diverse=True)
for para_phrase in para_phrases:
  print(para_phrase)
```

```python
#T5 Model
tokenizer = AutoTokenizer.from_pretrained("Vamsi/T5_Paraphrase_Paws")  
model = AutoModelForSeq2SeqLM.from_pretrained("Vamsi/T5_Paraphrase_Paws")
```

```python
#Paraphraseamos con el T5
#No sirve, mal modelo
#https://github.com/ceshine/finetuning-t5/tree/master/paraphrase

encoding = tokenizer.encode_plus(phrase,pad_to_max_length=True, return_tensors="pt")
input_ids, attention_masks = encoding["input_ids"], encoding["attention_mask"]

outputs = model.generate(
    input_ids=input_ids, attention_mask=attention_masks,
    max_length=70,
    do_sample=True,
    top_k=120,
    top_p=0.95,
    early_stopping=True,
    num_return_sequences=4
)

for output in outputs:
    line = tokenizer.decode(output, skip_special_tokens=True,clean_up_tokenization_spaces=True)
    print(line)
```

```python
#PEGASUS TRANSFORMER
#Es un meh, peor que el parrot pero mejor que el T5
#https://huggingface.co/docs/transformers/model_doc/pegasus#transformers.PegasusForConditionalGeneration
#https://arxiv.org/abs/1912.08777
from transformers import *
model = PegasusForConditionalGeneration.from_pretrained("tuner007/pegasus_paraphrase")
tokenizer = PegasusTokenizerFast.from_pretrained("tuner007/pegasus_paraphrase")
```

```python
def get_paraphrased_sentences(model, tokenizer, sentence, num_return_sequences=5, num_beams=5):
  # tokenize the text to be form of a list of token IDs
  inputs = tokenizer([sentence], truncation=True, padding="longest", return_tensors="pt")
  # generate the paraphrased sentences
  outputs = model.generate(
    **inputs,
    num_beams=num_beams,
    num_return_sequences=num_return_sequences,
  )
  # decode the generated sentences using the tokenizer to get them back to text
  return tokenizer.batch_decode(outputs, skip_special_tokens=True)
```

```python
get_paraphrased_sentences(model, tokenizer, phrase, num_beams=5, num_return_sequences=4)
```

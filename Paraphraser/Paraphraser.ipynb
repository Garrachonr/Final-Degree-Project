```python
#Importar librerias y tal

from parrot import Parrot
import torch
import warnings
warnings.filterwarnings("ignore")
```

```python
#Generar una semilla para la recreación de las frases
def random_state(seed):
  torch.manual_seed(seed)
  if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)

random_state(1234)
```

```python
#Cargar el modelo
parrot = Parrot(model_tag="prithivida/parrot_paraphraser_on_T5", use_gpu=False)
```

```python
#Frase a parafrasear, es un array de frases, por lo que se pueden meter más de una porque despues se recorre ese array
phrases = ["Some concepts or classes will come to you more naturally, so you can study those more quickly"]
```

```python
#Parafreamos y sacamos 4 ejemplos
for phrase in phrases:
  print("-"*100)
  print("Input_phrase: ", phrase)
  print("-"*100)
  para_phrases = parrot.augment(input_phrase=phrase, max_return_phrases = 5, do_diverse=True)
  for para_phrase in para_phrases:
   print(para_phrase)
```

```python
#Create Dataset
data = {"Title" : ["Titulo prueba"], "Text": ["Text prueba"], "Topic": ["Topic prueba"]}
post = pd.DataFrame(data)
post.to_excel("/content/drive/MyDrive/Trabajo Fin de Grado/Final-Degree-Project/Wrappers/Wiki.xlsx")
```

```python
#STUDY
how_to = HowTo("https://www.wikihow.com/Study")

for step in how_to.steps:
  dat = step.as_dict()
  prueba = {"Title":dat["summary"], "Text": dat["description"], "Topic": "Study"}
  post = post.append(prueba, ignore_index=True)
  # print(dat)

#FOCUS
how_to = HowTo("https://www.wikihow.com/Focus")

for step in how_to.steps:
  dat = step.as_dict()
  prueba = {"Title":dat["summary"], "Text": dat["description"], "Topic": "Focus"}
  post = post.append(prueba, ignore_index=True)
  # print(dat)

#LEARN
how_to = HowTo("https://www.wikihow.com/Learn")

for step in how_to.steps:
  dat = step.as_dict()
  prueba = {"Title":dat["summary"], "Text": dat["description"], "Topic": "Learn"}
  post = post.append(prueba, ignore_index=True)
  # print(dat)

#RELAX
how_to = HowTo("https://www.wikihow.com/Relax")

for step in how_to.steps:
  dat = step.as_dict()
  prueba = {"Title":dat["summary"], "Text": dat["description"], "Topic": "Relax"}
  post = post.append(prueba, ignore_index=True)
  # print(dat)
```

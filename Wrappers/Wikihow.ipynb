```python
import pandas as pd
from pywikihow import RandomHowTo, HowTo
from transformers import pipeline

#Resume
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
#WikiHow
how_to = HowTo("https://www.wikihow.com/Study")
```

```python
#Read data
#post = pd.read_excel("WikiHow.xlsx", index_col=0)
```

```python
#Create Dataset
data = {"Title" : ["Titulo prueba"], "Summary": ["Summary prueba"], "Resume": ["Resume prueba"], "ResumeShorter": ["ResumeShorter prueba"]}
post = pd.DataFrame(data)
post.to_excel("WikiHow.xlsx")
```

```python
for step in how_to.steps:
  dat = step.as_dict()
  resume = summarizer(dat["description"], max_length=150, min_length=50, do_sample=False)
  resume2 = summarizer(dat["description"], max_length=80, min_length=25, do_sample=False)
  prueba = {"Title":dat["summary"], "Summary": dat["description"], "Resume": resume, "ResumeShorter": resume2}
  post = post.append(prueba, ignore_index=True)
  # print(dat)
```

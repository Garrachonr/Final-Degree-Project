```python
#Pagina caida

import requests
import json


def jokes(f):
  data = requests.get(f)
  tt = json.loads(data.text)
  return tt

f = r"https://official-joke-api.appspot.com/random_ten"
a = jokes(f)

for i in (a):
  print(i["type"])
  print(i["setup"])
  print(i["punchline"], "\n")
```

```python
##DOC: https://github.com/leet-hakker/JokeAPI-Python#readme

from jokeapi import Jokes # Import the Jokes class
import asyncio

def print_joke():
    j = Jokes()  # Initialise the class
    joke = j.get_joke(lang="en", category=['pun', "misc"])
    # Retrieve a random joke
    print(joke)
    if joke["type"] == "single": # Print the joke
        print(joke["joke"])
    else:
        print(joke["setup"])
        print(joke["delivery"])

print_joke()
```

```python
#BEutifull Soap
#https://cinnamonxi.com/web-scraping-with-python-beautiful-soup/
#Import libraries
import pandas as pd
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen

#specify URL
url = "http://www.laughfactory.com/jokes/clean-jokes/"

#get the HTML from the page and assign it to the html variable
html = urlopen(url)

#parse the HTML and assign it to the variable 'soup'
soup = bs(html.read(), 'html.parser')

#get all jokes, authors, likes and dislikes then append to a list
authors = []
all_jokes = []
likes = []
dislikes = []
for jokes in soup.find_all("div", class_ = "jokes"):
    author = jokes.find("span", class_ = "joke-publisher").text
    authors.append(author)

    joke = jokes.find("div", class_ = "joke-text").text
    all_jokes.append(joke.strip())

    like = jokes.find("a", class_ = "like").text
    likes.append(int(like))

    dislike = jokes.find("a", class_ = "dislike").text
    dislikes.append(int(dislike))

#display the texts
display(authors)
display(all_jokes)
display(likes)
display(dislikes)
```

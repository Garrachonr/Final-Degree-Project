```python
#import model
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
#Create Dataset
import pandas as pd
data = {"Fact" : ["Fact prueba"], "Topic1": ["Topic1 prueba"], "Topic2": ["Topic2 prueba"]}
post = pd.DataFrame(data)
post.to_excel("TopicTrial.xlsx")
```

```python
#Read facts
data = pd.read_excel("/content/drive/MyDrive/Trabajo Fin de Grado/TFG/MongoDB/Reddit.xlsx", index_col=0)
```

```python
sequence_to_classify = "Frogs can't vomit. Instead, they eject their entire stomach out of their mouths and empty it using their hands."
candidate_labels = ['technology', 'science', 'health', 'economic', "maths", "politics", "chemistry", "animals"]
classifier(sequence_to_classify, candidate_labels)
```

```python
candidate_labels = ['technology', 'science', 'health', 'economic', "maths", "politics", "chemistry", "animals"]
for x in range(len(data)):
  fact = data.loc[x, "RedditFact"]
  topic = classifier(fact, candidate_labels)
  prueba = {"Fact": fact, "Topic1":(topic["labels"][0]+" "+str(topic["scores"][0])), "Topic2":(topic["labels"][1]+" "+str(topic["scores"][1]))}
  post = post.append(prueba, ignore_index=True)
```

```python
#SAVE
post.to_excel("TopicTrial.xlsx")
```
